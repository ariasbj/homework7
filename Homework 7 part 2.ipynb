```python
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
planet_data = sns.load_dataset('planets')
planet_data
```

```python
planet_data['Orbital Period'] = planet_data["orbital_period"]
planet_data['Mass'] = planet_data['mass']
```

```python
sns.scatterplot(data = planet_data, x = 'Orbital Period', y = 'Mass', hue = 'method')
plt.legend(fontsize = 'small')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data = planet_data, x = 'year', hue = 'method')
plt.xticks(rotation = 45)
plt.xlabel("Year")
plt.ylabel("Number of Planets Discovered")
plt.title("Number of Exoplanets Discovered by Year")
plt.tight_layout()
plt.show()
```
